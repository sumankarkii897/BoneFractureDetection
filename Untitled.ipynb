```python
import shutil
import glob
```

```python
base_dir=r"images"
```

```python
# Collect all fractured and nonfractured image
fractured_paths=glob.glob(os.path.join(base_dir,"*",'Fractured','*'))
```

```python
nonfractured_paths=glob.glob(os.path.join(base_dir,"*","Non-Fractured","*"))
```

```python
# combined Target Folder set the paths to new folders where all fractured and nonfractured images will be copied.
fractured_target=os.path.join(base_dir,"fracturedAll")
nonfractured_target=os.path.join(base_dir,"nonfracturedAll")
```

```python
# Create folders if they don't exit
os.makedirs(fractured_target,exist_ok=True) # exist_ok=True avoids errors if they already exist.
os.makedirs(nonfractured_target,exist_ok=True)
```

```python
# function to copy images
def copyImage(image_paths,target_dir):
    for path in image_paths:
        filename=os.path.basename(path) # get file name only
        base,ext=os.path.splitext(filename) # split the file name and extension of the file
        dst=os.path.join(target_dir,filename) # full destination path
        # For avoiding duplicate file name adding _1,_2 to make the file unique
        count=1
        while os.path.exists(dst):
            dst=os.path.join(target_dir,f"{base}_{count}{ext}")
            count+=1
        shutil.copy2(path,dst) # copy the file with metadata(creation date,etc)
    print("Image copied sucessfully ...")
```

```python
copyImage(fractured_paths,fractured_target)
copyImage(nonfractured_paths,nonfractured_target)
```
