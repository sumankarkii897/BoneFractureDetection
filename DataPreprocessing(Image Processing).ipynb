```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# load dataset
data = pd.read_csv(r"image_labels_path.csv")
data.head()
```

```python
# define Image Processing
def image_processing(img_path):
    img=cv2.imread(img_path)
    # plt.imshow(img)
    # convert to rgb
    img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    # resize to 100*100
    img_res=cv2.resize(img_rgb,(100,100))
    
    # convert to numpy arr
    img_ary=np.array(img_res).flatten()
    print(img_ary)
    # convert to dataframe
    img_df=pd.DataFrame(img_ary).T
    return img_df
    # plt.show()
```

```python
final_df=pd.DataFrame()
```

```python
# for feature in data.columns:
#     for img in data[feature]:
#         ret_df=image_processing(img)
#         final_df=pd.concat([final_df,ret_df])
for img in data["Fractured"]:
    df_x=image_processing(img)
    final_df=pd.concat([final_df,df_x])
final_df["label"]="Fractured"
```

```python
image_processing(data["Fractured"][0])
```

```python
final_df.index=range(len(final_df))
```

```python
final_dfx=pd.DataFrame()
```

```python
for img in data["Non-Fractured"]:
    df_x=image_processing(img)
    final_dfx=pd.concat([final_df,df_x])
final_dfx["label"]="Non-Fractured"
```

```python
# concating final_df and final_dfx 
final_images_df=pd.concat([final_df, final_dfx])
```
